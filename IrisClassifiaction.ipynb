```python
#selecting specific columns for the Training the Model using iloc that has sliced the Dataset
x=df.iloc[:,1:5]
y=df.iloc[:,5]
```

```python
x # Numerical data for Training the Model
```

Here we are using the sklearn library's train_test_split  function to split our data for Training

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
x_train.shape
```

here ,

I have used Logistic regression to classify the species

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train) # fitting the training values into the model for prdiction
```

```python
y_pred=model.predict(x_test) #making prediction from the test Dataset
```

```python
y_pred
pred=pd.DataFrame([y_pred])
pred
```

To check the Accuracy of the Model

I am using Confusion matrix and accuracy_score functions which are available in sklearn library

```python
from sklearn.metrics import accuracy_score,confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```
